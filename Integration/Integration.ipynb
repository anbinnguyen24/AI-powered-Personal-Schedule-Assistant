```python
import os
from langchain_nvidia_ai_endpoints import ChatNVIDIA

# 1. Đảm bảo bạn đã set API Key của NVIDIA
# Bỏ comment và thay bằng key thật của bạn nếu chưa set trong biến môi trường
# os.environ["NVIDIA_API_KEY"] = "nvapi-..."

def print_nvidia_models():
    # 2. Gọi hàm lấy toàn bộ danh sách mô hình từ NVIDIA API Catalog
    all_models = ChatNVIDIA.get_available_models()

    print(f"🔥 TỔNG SỐ MÔ HÌNH ĐƯỢC HỖ TRỢ: {len(all_models)}")
    print("=" * 50)

    tool_supported_models = []

    # 3. Duyệt qua từng mô hình và in ra thông tin
    for model in all_models:
        print(f"🔹 Tên Model (ID): {model.id}")

        # Kiểm tra xem mô hình có hỗ trợ Tool Calling không (cách này an toàn tránh lỗi thiếu thuộc tính)
        is_tool_supported = getattr(model, 'supports_tools', False)

        if is_tool_supported:
            print("   ✅ CÓ hỗ trợ Tool Calling")
            tool_supported_models.append(model.id)
        else:
            print("   ❌ KHÔNG hỗ trợ Tool Calling")

        print("-" * 40)

    # 4. In riêng danh sách các mô hình "chân ái" cho hệ thống Agent của bạn
    print("\n🎯 DANH SÁCH CÁC MÔ HÌNH CHUYÊN DÙNG CHO AGENT (HỖ TRỢ TOOL):")
    for tm in tool_supported_models:
        print(f" - {tm}")

if __name__ == "__main__":
    print_nvidia_models()
```

```python
import os
import numpy as np
from langchain_huggingface.embeddings import HuggingFaceEndpointEmbeddings

def cosine_similarity(vec1, vec2):
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))


# Danh sách dữ liệu mẫu
documents = [
    "Laptop học lập trình và trí tuệ nhân tạo, RAM lớn, chạy ổn định.",
    "Điện thoại có camera đẹp, pin lâu và màn hình sắc nét.",
    "Tai nghe không dây chống ồn, phù hợp học tập và làm việc.",
    "Máy tính bảng gọn nhẹ để đọc tài liệu và ghi chú trên lớp.",
    "Khóa học Python cơ bản dành cho sinh viên công nghệ thông tin.",
    "Khóa học machine learning giúp hiểu mô hình dự đoán dữ liệu."
]

# Khởi tạo embedding model
embeddings = HuggingFaceEndpointEmbeddings(
    model="sentence-transformers/all-mpnet-base-v2",
    task="feature-extraction",
)

# Tạo embedding cho toàn bộ documents
doc_vectors = embeddings.embed_documents(documents)

# Nhập truy vấn
query = input("Nhập câu cần tìm: ")

# Tạo embedding cho query
query_vector = embeddings.embed_query(query)

# Tính độ tương đồng
scores = []
for i, doc_vector in enumerate(doc_vectors):
    score = cosine_similarity(query_vector, doc_vector)
    scores.append((documents[i], score))

# Sắp xếp giảm dần theo điểm
scores.sort(key=lambda x: x[1], reverse=True)

# In kết quả tốt nhất
best_doc, best_score = scores[0]
print("\nKẾT QUẢ TÌM KIẾM:")
print("Câu giống nhất:", best_doc)
print("Điểm tương đồng:", round(best_score, 4))

# In top 3 nếu muốn
print("\nTOP 3 KẾT QUẢ:")
for doc, score in scores[:3]:
    print(f"- {doc} | score = {score:.4f}")
```

```python
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document
import shutil, os

# 4) Xóa DB cũ nếu có để tránh chạy lại bị trùng dữ liệu
persist_dir = "chroma_demo_db"
if os.path.exists(persist_dir):
    shutil.rmtree(persist_dir)

# 5) Embedding model miễn phí
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# 6) Dữ liệu mẫu
docs = [
    Document(
        page_content="LangChain là framework giúp xây dựng ứng dụng dùng LLM dễ dàng hơn.",
        metadata={"topic": "langchain", "level": "basic"}
    ),
    Document(
        page_content="Chroma là vector database dùng để lưu embeddings và tìm kiếm ngữ nghĩa.",
        metadata={"topic": "chroma", "level": "basic"}
    ),
    Document(
        page_content="Embeddings biến văn bản thành vector số để máy có thể so sánh độ giống nhau.",
        metadata={"topic": "embedding", "level": "basic"}
    ),
    Document(
        page_content="Retriever giúp lấy ra các tài liệu liên quan nhất cho câu hỏi của người dùng.",
        metadata={"topic": "retriever", "level": "basic"}
    ),
    Document(
        page_content="FAISS cũng là một vector store phổ biến cho bài toán tìm kiếm tương tự.",
        metadata={"topic": "faiss", "level": "extra"}
    ),
]

# 7) Tạo vector store
vector_store = Chroma(
    collection_name="demo_langchain_chroma",
    embedding_function=embeddings,
    persist_directory=persist_dir
)

# 8) Thêm documents
ids = [f"doc_{i}" for i in range(len(docs))]
vector_store.add_documents(documents=docs, ids=ids)

print("✅ Đã lưu", len(docs), "documents vào Chroma")

# 9) Semantic search
query = "Vector database dùng để lưu embedding là gì?"
results = vector_store.similarity_search(query, k=3)

print("\n" + "="*70)
print("1) KẾT QUẢ SEMANTIC SEARCH")
print("Query:", query)
print("="*70)

for i, doc in enumerate(results, 1):
    print(f"\nTop {i}")
    print("Nội dung :", doc.page_content)
    print("Metadata :", doc.metadata)

# 10) Search có filter metadata
filtered_results = vector_store.similarity_search(
    "Cái gì hỗ trợ lấy tài liệu liên quan?",
    k=3,
    filter={"topic": "retriever"}
)

print("\n" + "="*70)
print("2) KẾT QUẢ SEARCH CÓ FILTER (topic='retriever')")
print("="*70)

for i, doc in enumerate(filtered_results, 1):
    print(f"\nTop {i}")
    print("Nội dung :", doc.page_content)
    print("Metadata :", doc.metadata)

# 11) Dùng retriever
retriever = vector_store.as_retriever(search_kwargs={"k": 2})
retrieved_docs = retriever.invoke("Embeddings để làm gì?")

print("\n" + "="*70)
print("3) KẾT QUẢ RETRIEVER")
print("="*70)

for i, doc in enumerate(retrieved_docs, 1):
    print(f"\nTop {i}")
    print("Nội dung :", doc.page_content)
    print("Metadata :", doc.metadata)

print("\n🎯 Gợi ý tự luyện:")
print("- Thêm docs mới về RAG hoặc LLM")
print("- Đổi query để xem semantic search thay đổi ra sao")
print("- Thử filter={'level': 'basic'}")
print("- Đổi k=1, k=2, k=4 để quan sát khác biệt")
```

```python

from langchain_unstructured import UnstructuredLoader

# Tạo file mẫu
sample_text = """
LangChain là framework hỗ trợ xây dựng ứng dụng với mô hình ngôn ngữ lớn.

Unstructured API giúp tách nội dung từ tài liệu và trả về các document có metadata.

Khi kết hợp LangChain với Unstructured API, chúng ta có thể đọc tài liệu,
biến tài liệu thành Document, rồi dùng tiếp cho RAG hoặc chatbot hỏi đáp tài liệu.
"""

file_path = "demo_unstructured_api.txt"
with open(file_path, "w", encoding="utf-8") as f:
    f.write(sample_text)

# Dùng Unstructured API (không local)
loader = UnstructuredLoader(
    file_path=file_path,
    partition_via_api=True,
    strategy="fast",
    chunking_strategy="basic",
)

docs = loader.load()

print("✅ Số lượng docs:", len(docs))

print("\n" + "="*70)
print("1) DOCUMENT ĐẦU TIÊN")
print("="*70)
print(docs[0])

print("\n" + "="*70)
print("2) NỘI DUNG + METADATA")
print("="*70)
for i, doc in enumerate(docs, 1):
    print(f"\n--- Document {i} ---")
    print("Nội dung :", doc.page_content[:500])
    print("Metadata :", doc.metadata)
```
